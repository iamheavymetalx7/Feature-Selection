## feature selection with correlation
dropping the features which are highly correlated

```python
from sklearn.datasets import load_boston
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = load_boston()
df=pd.DataFrame(data.data, columns = data.feature_names)
##all independentfeatures
df["MEDV"]=data.target
```

```python
data.feature_names
```

```python
df.head()
```

if a independent feature is highly correlated with dependent features, we need not remove them, as they play an important role for traing machine learning models.

but if we have two or more dependent feature are highly corelated (>90%) we prefer to just use one of them


```python
X=df.drop("MEDV",axis=1)
y=df["MEDV"]
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test =  train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=0

)
X_train.shape, X_test.shape
```

```python
cor=X_train.corr()
## peaarson corr ranges between -1 to +1
cor
```

```python
import seaborn as sns
## USING PEARSON CORRELATION
plt.figure(figsize=(12,10))
cor=X_train.corr()
sns.heatmap(cor,annot=True, cmap=plt.cm.CMRmap_r)
plt.show()
```

```python
# with the following function we can select highly correlated features
# it will remove the first feature that is correlated with anything other feature

def correlation(dataset, threshold):
    col_corr = set()  # Set of all the names of correlated columns
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: # we aret interested in absolute coeff value
                colname = corr_matrix.columns[i]  # getting the name of column
                col_corr.add(colname)
    return col_corr

    ## note that we have used abs(corr_matrix) but negatively correlated features also very important
```

```python
corr_features = correlation(X_train,0.7)
len(set(corr_features))
#len(corr_features)


## to avoud overfitting we will not do this in X_test, we directly remove the same columns from X_test, X_train both
```

```python
corr_features
```

```python
X_train.drop(corr_features,axis=1)
X_test.drop(corr_features,axis=1)
```
